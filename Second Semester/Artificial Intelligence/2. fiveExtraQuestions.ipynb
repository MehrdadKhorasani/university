```python
# name of companies that germans buy something from:
df[df['Language'] == "de"]['Company'].unique()
```

```python
# number of purchases above 90 by engineers:
def isEngineer(title):
  if 'engineer' in title.lower():
    return True
  return False
sum(df[df['Job'].apply(lambda x: isEngineer(x))]['Purchase Price'] > 90)
```

```python
# google lovers:
def isChrome(browser):
  if 'chrome' in browser.lower(): return True
  return False
def isGmail(email):
  if 'gmail' in email.lower(): return True
  else: return False
sum(df[df['Browser Info'].apply(lambda b: isChrome(b))]['Email'].apply(lambda e: isGmail(e)))
```

```python
# purchases between 16 and 32:
sum(df[df['Purchase Price'] > 16]['Purchase Price'] < 32)
```

```python
# sum of the addresses that Purchase more than once!
sum(df['Address'].value_counts() > 1)
```
